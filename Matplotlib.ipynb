```python
# Side by Side bar chart
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values1 = [5, 7, 3, 8, 4]  # Group 1
values2 = [6, 5, 7, 6, 5]  # Group 2
values3 = [7, 4, 2, 3, 7] # Group 3

bar_width=0.2
x=np.arange(len(categories))
# x=1
# 1-0.2=0.8
plt.bar(x-bar_width/3, values1, width=bar_width,color='lightblue', label='Group 1')
plt.bar(x+bar_width/3, values2, width=bar_width,color='orange', label='Group 2', alpha=0.7)
plt.bar(x+bar_width, values3, width=bar_width,color='blue', label='Group 3', )
plt.plot(x-bar_width/3, values1,color='lightblue')
plt.plot(x+bar_width/3, values2,color='orange')
plt.plot(x+bar_width, values3,color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x, categories) # replace numeric values of x aaxis by the values of categories
plt.title('Bar Plot with Multiple Groups')
plt.legend(loc="upper right", bbox_to_anchor=(1.2,1))

plt.show()
```

```python
# Pie Chat
import matplotlib.pyplot as plt
manufacturers=['Samsung','Apple','Huawei','Xiaomi','Others']
market_share=[30, 25, 18, 12, 15]
colors=['Red','green','yellow','orange','blue']
plt.pie(market_share,labels=manufacturers,autopct='%1.1f%%',textprops={'fontsize':10,'fontweight':'bold', 'color':'black'}) # Persentage use %1.1f%%
plt.legend(loc='upper right', bbox_to_anchor=(1.5,1)) # heading block side
plt.title('Smartphone Market Contribution Share Analysis')
plt.show()
```

```python
# Donut Chart
import matplotlib.pyplot as plt
manufacturers=['Samsung','Apple','Huawei','Xiaomi','Others']
market_share=[30, 25, 18, 12, 15]
autopct_format=lambda p:f'{p*sum(market_share)/100:.1f}'
# 30*100/100= 30.0
# 25*100/100= 25.0
# 18*100/100= 18.0
# 12*100/100= 12.0
# 15*100/100= 15.0
plt.pie(market_share,labels=manufacturers,autopct=autopct_format,textprops={'fontsize':10,'fontweight':'bold', 'color':'black'})
circle=plt.Circle((0,0),0.6,color='white')
fig=plt.gcf()
fig.gca().add_artist(circle)
plt.title('Smartphone Market Contribution Share Analysis')
plt.show();
```

```python
# Create Pie Chat
import matplotlib.pyplot as plt
# Data
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]
colors = ['gray', 'silver', 'olive', 'teal']
# Create Pie Chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, textprops={'fontsize':10, 'fontweight':'bold'})
# Title
plt.title('Pie Chart Example')
# Show Plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]
colors = ['gray', 'silver', 'olive', 'teal']

# Create Pie Chart (Donut Chart) without percentages
plt.pie(sizes, labels=labels,autopct=autopct_format, colors=colors, textprops={'fontsize':10, 'fontweight':'bold'})

# Add Donut Hole
circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(circle)

# Title
plt.title('Donut Chart Example')

# Show Plot
plt.show()
```

# A Histogram is a type of chart that shows how often different values appear in a dataset.

Think of it like this:
Imagine you have 50 students' test scores ranging from 1 to 100. Instead of listing all the scores, you can group them into ranges (like 0-10, 10-20, 20-30, etc.). The histogram tells you how many students got scores in each range by drawing bars.

How it Works:
Collect the data (e.g., students' scores).
# Group the data into "bins" (ranges) (e.g., 0-10, 10-20, etc.).

```python
# Histogram Chat
import matplotlib.pyplot as plt
data=[1,2,2,3,5,6,6,6,7,8,9,9,10,11,12,13,13,14]
plt.hist(data, bins=7, color="olive", edgecolor="black")
# 0-20 20-40 40-60 60-80 80-100
# 0-25 25-50 50-75 75-100
#
plt.xlabel("Values")
plt.ylabel("Frequency of Distribution")
plt.title("Histogram")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
# Create a subplot with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
# Plot Math Scores
axs[0].bar(students, math_scores, color='olive')
axs[0].set_title('Math Scores')
axs[0].set_xlabel('Students')
axs[0].set_ylabel('Score')
# Plot Science Scores
axs[1].bar(students, science_scores, color= 'teal')
axs[1].set_title('Science Scores')
axs[1].set_xlabel('Students')
axs[1].set_ylabel('Score')
# Adjust spacing between subplots
plt.tight_layout()
# Show the subplot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
english_scores = [88, 96, 84, 90, 73]
cs_scores = [70, 85, 95, 85, 75]
# Create a subplot with two subplots (2 row, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(20, 8))
# Plot Math Scores
axs[0][0].bar(students, math_scores, color='olive')
axs[0][0].set_title('Math Scores')
axs[0][0].set_xlabel('Students')
axs[0][0].set_ylabel('Score')
# Plot Science Scores
axs[0][1].bar(students, science_scores, color='silver')
axs[0][1].set_title('Science Scores')
axs[0][1].set_xlabel('Students')
axs[0][1].set_ylabel('Score')
axs[1][0].bar(students, english_scores, color='teal')
axs[1][0].set_title('English Scores')
axs[1][0].set_xlabel('Students')
axs[1][0].set_ylabel('Score')
axs[1][1].bar(students, cs_scores, color='gray')
axs[1][1].set_title('Computer Scores')
axs[1][1].set_xlabel('Students')
axs[1][1].set_ylabel('Score')
# Adjust spacing between subplots
plt.tight_layout()
# Show the subplot
plt.show()
```

# Qustion 1 : 
Create a histogram to represent the distribution of product prices in a retail
store and give your conclusion.
Input:
product_prices = [24.99, 34.99, 49.99, 64.99, 39.99, 54.99, 79.99, 99.99, 29.99, 44.99,
59.99, 69.99, 84.99, 109.99, 119.99, 89.99, 74.99, 124.99, 69.99, 54.99] 

```python
import matplotlib.pyplot as plt
product_prices = [24.99, 34.99, 49.99, 64.99, 39.99, 54.99, 79.99, 99.99, 29.99, 44.99, 59.99, 69.99, 84.99, 109.99, 119.99, 89.99, 74.99, 124.99, 69.99, 54.99]

# Define bins (Price Ranges)
bins = [20, 40, 60, 80, 100, 120, 140]  # Creating price intervals

# Create Histogram
plt.hist(product_prices, bins=bins, color='olive', edgecolor='black', alpha=0.7)

# Labels & Title
plt.xlabel('Price Range ', fontsize=12, fontweight='bold')
plt.ylabel('Number of Products', fontsize=12, fontweight='bold')
plt.title('Distribution of Product Prices in a Retail Store', fontsize=14, fontweight='bold')

plt.xticks(bins)  # Ensure proper bin labels

# Show Plot
plt.show()
```

The histogram shows that most products are cost between 40 and 80, making mid-range prices the most common. There are fewer high-end products above 100 

# Most products are priced between 40 and 80.
# Few products are priced above 100.

```python
import numpy as np
import matplotlib.pyplot as plt

square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])  # in $1000s

# Create Scatter Plot
plt.figure(figsize=(8, 10))
plt.scatter(square_footage, selling_prices, color='blue', marker='o', label='Data Points')

# Labels & Title
plt.xlabel('House Size (Square Footage)', fontsize=12, fontweight='bold')
plt.ylabel('Selling Price ', fontsize=12, fontweight='bold')
plt.title('House Size vs. Selling Price', fontsize=14, fontweight='bold')
plt.legend()
plt.show()
```

Bigger houses sell for more money. The scatter plot shows that as the house size increases, the selling price also goes up

```python
import numpy as np
import matplotlib.pyplot as plt

# Data: Sales for 12 months
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000,
23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000,
25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000,
19000, 20000, 21000, 22000, 23000])


plt.figure(figsize=(10, 6))  # create plot size

plt.plot(months, electronics_sales, marker='o', linestyle='-', color='blue', label='Electronics')
plt.plot(months, clothing_sales, marker='s', linestyle='-', color='green', label='Clothing')
plt.plot(months, home_garden_sales, marker='^', linestyle='-', color='red', label='Home & Garden')
plt.plot(months, sports_outdoors_sales, marker='d', linestyle='-', color='purple', label='Sports & Outdoors')
# Labels & Title
plt.xlabel('Months', fontsize=12, fontweight='bold')
plt.ylabel('Sales', fontsize=12, fontweight='bold')
plt.title('Sales Performance by Product Category', fontsize=14, fontweight='bold')

plt.legend()
plt.show()
```

Electronics sells the most keeps growing fast. Other products like Clothing, Home & Garden, and Sports & Outdoors are also growing but more slowly.
